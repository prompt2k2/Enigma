### Remove all Postcode without Coordinates

```python
df=dfs[dfs.Latitude !='NaN']
df.shape
```

```python
df
```

#### Plot the Marker with Coordinates

```python
for lat, long, bor, nei in zip(df['Latitude'],df['Longitude'], df['Borough'], df['Neighborhood']):
    label='{},{}'.format('neighborhood','borough')
    label=folium.Popup(label,perse_html=True)
    folium.CircleMarker(
        [lat,long], radius=5,
        popup=label, color='red', fill=True,
        fill_color='r', 
        fill_opacity=0.6, 
        parse_html=False).add_to(Toronto_Map)
Toronto_Map
```
